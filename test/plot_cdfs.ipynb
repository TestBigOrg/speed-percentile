```python
import numpy as np
from collections import Counter
import json, os, shutil

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
sns.set(font='Avenir', style='white')
```

```python
size = 100000
dir = 'simulate_' + str(int(size)) + '/'

with open(dir + 'cdfs.json') as f:
    data = json.load(f)
```

```python
dists = sorted(list(set(data.keys()) - set(['ps'])))
```

```python
fig = plt.figure(figsize=(16,6))

i = 0
for dist in dists:
    i += 1
    ax = fig.add_subplot(1,3,i)
    
    for algorithm,speeds in data[dist]['speed'].iteritems():
        ax.plot(speeds, data['ps'], label=algorithm)
    ax.set_title(dist)
    ax.set_xlabel('speed')
    ax.set_ylabel('percentile')
    ax.legend(loc=2)

plt.suptitle('percentile speed', size=20)
plt.tight_layout()
plt.subplots_adjust(top=0.85)

plt.savefig(dir + 'cdfs.png', dpi=300)
```
